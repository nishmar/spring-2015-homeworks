## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here

#Add a new column with the date, ignoring time information
boston["date"] = boston.OPEN_DT.apply(lambda d: d.replace(hour=0, minute=0, second=0, microsecond=0))

#Find top 5 cases
top_5_cases = boston.CASE_TITLE.value_counts().head(5)

#Convert series to dataframe to access index name as a column
top_5_cases.to_csv('top_5_cases.csv', index=True)
top_5_cases = pd.read_csv('top_5_cases.csv', names= ['CASE_TITLE', 'COUNTS'])

#New boston df to only include the top 5 cases
boston_top = pd.DataFrame()

#Find case titles matching each case in the top 5 and add to new df
for i in range(0,5):
    boston_top = boston_top.append(boston[boston.CASE_TITLE == top_5_cases.CASE_TITLE[i]])

#New dataframe to store counts for top 5 cases per day
boston_top_per_day = pd.DataFrame()

#List of days
day_list = []

#Go through each day
for day, day_data in boston_top.groupby("date"):
    current_day = boston_top[boston_top.date == day]
    
    #Series of case titles and their counts of each day
    day_cases = current_day.CASE_TITLE.value_counts()
    
    #Store series in dataframe
    boston_top_per_day = boston_top_per_day.append([day_cases])
    
    #Append day to day list
    day_list.append(day)

#Add the date column to dataframe from collected list     
boston_top_per_day["DATE"] = day_list

#Set these dates as the index
boston_top_per_day = boston_top_per_day.set_index(["DATE"])

#Replace NaN values with 0
boston_top_per_day = boston_top_per_day.fillna(0)

#Plot line graph
boston_top_per_day.plot().set_ylabel("# Calls")
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here

#Filter SNOW column for negative quantities
snow_data_positive = snow_data[snow_data.SNOW>= 0]
snow_data_negative = snow_data[snow_data.SNOW < 0]

#Set all negative values to 0
snow_data_negative.SNOW = 0

#Include values back into original df
snow_data_positive = snow_data_positive.append(snow_data_negative)
snow_data = snow_data_positive

#Initialize empty lists
day_list = []
snow_list = []

#Group by date, average out snow fall totals over stations
for day, day_data in snow_data.groupby("DATE"):
    current_day = snow_data[snow_data.DATE == day]
    
    #Get mean and convert to cm
    snow_avg_cm = current_day.SNOW.mean()/ 10.0 
    
    #Add to list of days
    day_list.append(day)
    #Add to list of snow fall totals
    snow_list.append(snow_avg_cm)

#Create aggregate df
snow_totals = pd.DataFrame(columns=["DATE","SNOW_AVG"])
snow_totals["DATE"] = day_list
snow_totals["SNOW_AVG"] = snow_list

#Set date as index
snow_totals = snow_totals.set_index(["DATE"])

#Plot results
snow_totals.plot().set_ylabel("Snowfall in cm")

````

## Step 4. Matching trends

The last thing we will do is to join the snowfall dataframe with that of the top 5 categories of calls, per day. Don't forget to fill the NaN values with 0. You want to do an outer join (if there are calls for a day but no snowfall data, we want to create a 0 record). Save that to a dataframe called `calls_on_snow`. The first 5 columns should be the counts for the categories, and the 6th will be the snowfall data. **(10 pts)**

````python
#Outer join top 5 categories of calls, per day with snow totals
calls_on_snow = boston_top_per_day.join(snow_totals, how='outer')

#Replace any NaN values with 0
calls_on_snow = calls_on_snow.fillna(0)
````

Now, you can show the connection of the two trends (the calls and the snow) in a plot. I provide some sample code, but you should also try your own versions.

````python
fig, axes = plt.subplots(nrows=2,figsize=(12,5), sharex=True)

calls_on_snow.ix[:,0:5].plot(ax=axes[0])
calls_on_snow.ix[:,5].plot(ax=axes[1],label="snowfall (in cm)", color='gray')

axes[0].set_title("The effect of the winter snowstorms on Boston's 311 Hotline for 2015")
axes[0].set_ylabel("# Calls")
axes[1].set_ylabel("Snowfall in cm")
axes[1].set_xlabel("Date")
axes[0].xaxis.grid(False)
axes[1].xaxis.grid(False)
````
